```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub_df.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub_df.shape[1]), sub_df.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub_df.shape[1]), sub_df.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub_df.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

    #The coefficient shows the correlation between two variables. 
    0: no relationship
    0.1–0.3: weak relationship
    0.3–0.5: moderate relationship
    0.5–1.0: strong relationship

#Categorical Plot on Prices and LivingSpace

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['Median Household Income'], bins = 20, kde = True)
```

    Category
    < 100000
    100000 - 200000
    > 200000

```python
df.loc[df['Median Household Income'] < 100000, 'Income category'] = 'Low Income'
```

```python
df.loc[(df['Median Household Income'] >= 100000) & (df['Median Household Income'] < 200000), 'Income category'] = 'Medium Income'
```

```python
df.loc[df['Median Household Income'] >= 200000, 'Income category'] = 'High Income'
```

```python
df['Income category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(y ="Price", x="Median Household Income", hue="Income category", data=df)
```

    # when the Price is above 1.75 million, there are only high or medium income groups

```python
sns.histplot(df['Zip Code Population'], bins = 10, kde = True)
```

    Category
    < 20000 - Low Populated
    20000 - 50000 - Medium Populated
    >= 50000 - Highly Populated


```python
df.loc[df['Zip Code Population'] < 20000, 'ZipCode category'] = 'Low Populated'
```

```python
df.loc[(df['Zip Code Population'] >= 20000) & (df['Zip Code Population'] < 50000), 'ZipCode category'] = 'Medium Populated'
```

```python
df.loc[df['Zip Code Population'] >= 50000, 'ZipCode category'] = 'Highly Populated'
```

```python
df['ZipCode category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(y ="Price", x ="Zip Code Population", hue="ZipCode category", data=df)
```
